# 1) Generating and Plotting the elementary signals in DSP

## Impulse Signal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10)
n =  np.size(x)
y = np.zeros(n)
i = np.where(x==0)
y[i] = 1

plt.stem(x,y, use_line_collection=True)
plt.title("Impulse Signal")
plt.xlabel("Sample Number")
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()
```

## Unit Step Signal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10)
n =  np.size(x)
y = np.zeros(n)
i = np.where(x>=0)
y[i] = 1

plt.stem(x,y, use_line_collection=True)
plt.title("Unit Step Signal")
plt.xlabel("Sample Number")
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()
```

## Ramp Signal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10)
n =  np.size(x)
y = np.zeros(n)
i = np.where(x>=0)
y[i] = x[i]

plt.stem(x,y, use_line_collection=True)
plt.title("Ramp Signal")
plt.xlabel("Sample Number")
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()
```

## Exponential Signals

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10)
n = -1.2
y = n**x

plt.stem(x,y, use_line_collection=True)
plt.title("Exponential Signal")
plt.xlabel("Sample Number")
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()
```
